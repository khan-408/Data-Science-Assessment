```python
import requests
import json
import pandas as pd

def download_data(api_link):
    response = requests.get(api_link)
    if response.status_code == 200:
        return response.json()
    else:
        print("Failed to download data from the API.")
        return None

def extract_data(data):
    show_name = data['name']
    premiered = data['premiered']
    status = data['status']
    average_rating = data['rating']['average']
    genres = ", ".join(data['genres'])
    summary = data['summary']
    episodes = data['_embedded']['episodes']
    
    episode_data = []
    for episode in episodes:
        episode_name = episode['name']
        season = episode['season']
        episode_number = episode['number']
        airdate = episode['airdate']
        episode_data.append({
            'Episode Name': episode_name,
            'Season': season,
            'Episode Number': episode_number,
            'Airdate': airdate
        })
    
    show_data = {
        'Show Name': show_name,
        'Premiered': premiered,
        'Status': status,
        'Average Rating': average_rating,
        'Genres': genres,
        'Summary': summary
    }
    
    return pd.DataFrame([show_data]), pd.DataFrame(episode_data)

# API link
api_link = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Download data
data = download_data(api_link)

# Extract and display the required information as a dataframe
if data:
    show_df, episode_df = extract_data(data)
    print("Show Information:")
    print(show_df)
    print("\nEpisode Information:")
    print(episode_df)
```
