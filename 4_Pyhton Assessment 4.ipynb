### Write a program to download the data from the link given below and then read the data and convert the into the proper structure and return it as a CSV file.
Link - https://data.nasa.gov/resource/y77d-th95.json
Note - Write code comments wherever needed for code understanding.

Excepted Output Data Attributes
● Name of Earth Meteorite - string id - ID of Earth
● Meteorite - int nametype - string recclass - string
● mass - Mass of Earth Meteorite - float year - Year at which Earth
● Meteorite was hit - datetime format reclat - float recclong - float
● point coordinates - list of int


```python
import pandas as pd
import ast
from datetime import date
```

```python
import urllib

from urllib.request import urlopen

 

# import json

import json
from pandas.io.json import json_normalize

# store the URL in url as

# parameter for urlopen

url = "https://data.nasa.gov/resource/y77d-th95.json"

 

# store the response of URL

response = urlopen(url)

 

# storing the JSON response

# from url in data

data_json = json.loads(response.read())

 

df=json_normalize(data_json)
```

```python
df
```

```python
df.year
```

```python
df["year"]=df.year.str.split("T").str[0]
```

```python
df[df["name"]=='Tomakovka']
# checking from sample
```

```python
# ● Name of Earth Meteorite - string 
# id - ID of Earth Meteorite - int
# nametype - string 
# recclass - string
# ● mass - Mass of Earth Meteorite - float 
# year - Year at which Earth Meteorite was hit - datetime
#  reclat - float 
# recclong - float
# ● point coordinates - list of int
```

```python
df["id"].dtype
```

```python
df["id"]=df["id"].astype(int)
```

```python
df["id"].dtype
```

```python
df["mass"]=df["mass"].astype(float)
```

```python
df['year'].isnull().sum()
```

```python
df[df['year'].isnull()]
```

```python
# To Removing the null values in year column
# 37 is the index of the instace where the year is null 
df.drop(37,inplace=True)
```

```python
df['year'].isnull().sum()
```

```python
df['year'] = pd.to_datetime(df['year'],errors = 'coerce')
```

```python
df["reclat"]=df["reclat"].astype(float)
```

```python
df["reclong"]=df["reclong"].astype(float)
```

```python
df["reclong"].dtype
```

```python
df["geolocation.coordinates"].dtype
```

```python
df["geolocation.coordinates"] = df["geolocation.coordinates"].astype(str)
```

```python
df['geolocation.coordinates'].value_counts()
```

```python
df.drop(df[df['geolocation.coordinates'] == 'nan'].index,inplace=True)

```

```python
df['geolocation.coordinates'].value_counts()
```

```python
lst = []
for i in df['geolocation.coordinates']:
    i = ast.literal_eval(i)
    lst.append([int(round(x)) for x in i])
```

```python
lst
```

```python
df['geolocation.coordinates']=lst
```

```python
df
```

```python
df.to_csv("Nasa.csv",index=False,header=True)
```
